# Aim: Rearrange the order of columns and indices
The reformed ingredient properties will have most cooccuring ingredients and it's material properties at the top left, which will be easier for visualisation.

```python
df = pd.read_excel('lcooccurrence_ing_vs_matprop_v9.xlsx', index_col = [0])
print(df.shape)
df.head(10)
```

```python
# default columns order
collst = df.columns.tolist()
print(len(collst))
print(collst)
```

```python
coldict = {}
for col in collst:
    sumcol = df[col].sum()
    #rint(col.ljust(40),sumcol) 
    coldict[col] = int(sumcol)
```

### Sum total of the values of a particular column and the column name in a dictionary

```python
print(dict(coldict))
```

## Sorting the dictionary to get the heights valued column at the top

```python
ncoldict = dict(sorted(coldict.items(), key=lambda item: item[1],reverse = True))
print(len(ncoldict))
print(ncoldict)
```

```python
ncoldict
```

## `ncollst` is the required order of columns

```python
ncollst = []
for k,v in ncoldict.items():
    ncollst.append(k)
print(len(ncollst),ncollst)
```

## Rearranging columns and indices

```python
# rearranging columns
dff = df.reindex(columns = ncollst)
dff.head(15)
```

```python
dff['sum'] = dff.sum(axis =1)
```

```python
sumlst = dff['sum'].tolist()
print(len(sumlst),sumlst)
idxlst = dff.index.tolist()
print(len(idxlst),idxlst)
```

```python
idxdict = {}
for sval in range(len(sumlst)):
    idxdict[idxlst[sval]] = sumlst[sval]
print(len(idxdict))
```

```python
print(dict(idxdict))
```

## Sorting the list to get the hightest value at the top

```python
nidxdict = dict(sorted(idxdict.items(), key=lambda item: item[1],reverse = True))
print(len(nidxdict))
print(nidxdict)
```

```python
nidxlst = []
for k,v in nidxdict.items():
    nidxlst.append(k)
print(len(nidxlst),nidxlst)
```

## Rearranging index to get the most number of cooccurrences at the top left

```python
# rearrainging indices
redf = dff.reindex(nidxlst)
```

```python
print(redf.shape)
redf.drop('sum',axis = 1, inplace = True)
redf.shape
```
